```python
import numpy as np
import matplotlib.pyplot as plt

# Plate parameters
plate_separation = 0.1  # meters (distance between plates along z-axis)
plate_length = 0.2      # meters (length along y-axis)
V = 100                 # Voltage across plates (Volts)
d = plate_separation    # Distance between plates

# Field magnitude in between the plates (uniform field)
E_magnitude = V / d     # V/m

# Create grid for visualization (2D yz-plane)
y = np.linspace(-plate_length/2, plate_length/2, 30)
z = np.linspace(0, plate_separation, 40)
Y, Z = np.meshgrid(y, z)

# Initialize field vector components
Ey = np.zeros_like(Y)
Ez = -E_magnitude * np.ones_like(Z)  # Field points downward (−z)

# Log scaling for vector magnitudes (optional and clamped)
E_mags = np.sqrt(Ey**2 + Ez**2)
E_logscale = np.log1p(E_mags / E_mags.max())  # log(1 + x) scaling, 0–1 range

# Normalize and rescale vector directions
Ey_scaled = Ey * E_logscale
Ez_scaled = Ez * E_logscale

# Plotting
fig, ax = plt.subplots(figsize=(8, 6))
ax.quiver(Y, Z, Ey_scaled, Ez_scaled, color='blue', scale=1, scale_units='xy', angles='xy')

# Draw the capacitor plates
plate_thickness = 0.005
ax.add_patch(plt.Rectangle((-plate_length/2, plate_separation - plate_thickness), plate_length, plate_thickness,
                           color='red', alpha=0.9, label='Positive Plate (+)'))
ax.add_patch(plt.Rectangle((-plate_length/2, 0), plate_length, plate_thickness,
                           color='blue', alpha=0.9, label='Negative Plate (−)'))

# Formatting
ax.set_title("Electric Field in a Parallel Plate Capacitor (Side View)", fontsize=14)
ax.set_xlabel("y-axis (m)", fontsize=12)
ax.set_ylabel("z-axis (m)", fontsize=12)
ax.set_xlim(-plate_length/2, plate_length/2)
ax.set_ylim(-0.01, plate_separation + 0.01)
ax.grid(True)
ax.legend()
plt.tight_layout()
plt.show()
```
